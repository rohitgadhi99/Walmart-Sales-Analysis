# **DATA CLEANING**

```python
df.describe()
```

```python
df.info()
```

```python
# All duplicates
df.duplicated().sum()
```

```python
# To drop duplicates
df.drop_duplicates(inplace=True)

df.duplicated().sum()
```

```python
df.shape
```

```python
# No of null values
df.isnull().sum()
```

```python
# Remove/dropping all rows with missing records (here null values)
df.dropna(inplace=True)

#verify
df.isnull().sum()
```

```python
df.shape
```

```python
# To convert the 'unit_price' dtype from object into float - we got error because of the dollar sign
df['unit_price'].astype(float)
```

```python
df['unit_price'].str.replace('$', '')

df.head()
```

```python
# Replace $ with nothing and convert dtype
df['unit_price'] = df['unit_price'].str.replace('$', '').astype(float)
df.head()
```
